```python
import numpy as np
import matplotlib.pyplot as plt
```

1. Vazifa: $ f(x) = x^2 - 4x + 4 $ funksiyasini $ x $ qiymatlari uchun -10 va 10 oralig'ida chizing. Syujetni o'qlar va sarlavha uchun mos teglar bilan moslang.

```python
def f(x):
    return x**2 -4*x + 4

x = np.linspace(-10, 10, 10000)
y = f(x)

plt.plot(x, y, c='skyblue', label='$f(x)=x^2 -4x + 4$')
plt.title('Square Function')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.legend()
plt.show()
```

2. $ \sin(x) $ va $ \cos(x) $ ni 0 dan $ 2\pi $ gacha boʻlgan $ x $ qiymatlari uchun bir xil grafikda chizing. Ikki funktsiyani farqlash uchun turli xil chiziq uslublari, markerlar va ranglardan foydalaning. Afsona qo'shing.

```python
def cosinus(n):
    return np.cos(n*np.pi)

def sinus(n):
    return np.sin(n*np.pi)

x = np.linspace(0, 2, 100)
cos = cosinus(x)
sin = sinus(x)

plt.plot(x, cos, c='r', label='$f(x)=cos(x)$')
plt.plot(x, sin, c='y', label='$f(x)=sin(x)$')
plt.title('Square Function')
plt.xlabel('Cos(x)')
plt.ylabel('Sin(x)')
plt.legend()
plt.show()
```

3. Vazifa: 2x2 o'lchamdagi pastki chizmalar to'plamini yarating. Har bir kichik uchastkada chizma: $ \\ $ Yuqori chap: $ f(x) = x^3 \\ $ Yuqori oʻng: $ f(x) = \sin(x) \\ $ Pastki chap: $ f(x) = e^x  \\ $ Pastki oʻng: $ f(x) = \log(x+1) $ ($ x \geq 0 $ uchun) \\ $ Har bir uchastkani sarlavhalar, eksa yorliqlari va turli ranglar bilan moslashtiring.

```python
def f1(x):
    return x**3

def f2(x):
    return np.sin(x * np.pi)

def f3(x):
    return np.e ** x

def f4(x):
    return np.where(x >= 0, np.log10(x + 1), None)

n = np.linspace(-2, 2, 100)
x = np.arange(20)

plt.subplot(2, 2, 1)
plt.plot(x, f1(x), c='k', marker='*', linestyle='--',  label='$f(x)=x^3$')
plt.title('Kubik funksiya')
plt.xlabel('X')
plt.ylabel('$f(x)=x^3$')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(n, f2(n), marker='>', linestyle='-.', label='$f(x) = sin(x)$')
plt.title('Sinus funksiya')
plt.xlabel('X')
plt.ylabel('$f(x) = sin(x)$')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, f3(x), c='g', marker='+', linestyle='dotted', label='$f(x)=e^x$')
plt.title('Exponeneta funksiya')
plt.xlabel('X')
plt.ylabel('$f(x)=e^x$')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, f4(x), c='b', marker='o', linestyle=':', label='$f(x)=log(x + 1$')
plt.title('Logarfimik funksiya')
plt.xlabel('X')
plt.ylabel('$f(x)=log(x + 1$')
plt.legend()

plt.show()
```

5. Oddiy taqsimotdan tanlab olingan 1000 qiymatdan iborat tasodifiy ma'lumotlar to'plamini yarating (o'rtacha = 0, std = 1). Ma'lumotlarning gistogrammasini 30 ta quti bilan tuzing. Sarlavha va o'q teglarini qo'shing. Alfa parametri yordamida barlarning shaffofligini sozlang.

6. $ f(x, y) = \cos(x^2 + y^2) $ funktsiyasi uchun $ x $ va $ y $ qiymatlari -5 dan 5 gacha bo'lgan 3D sirt syujetini yarating. Tegishli rang xaritasidan foydalaning. va rang paneli qo'shing. O'qlar va sarlavha uchun tegishli teglarni o'rnating.

```python
x = np.arange(0, 1, 0.01)
y = np.arange(0, 1, 0.01)

def f(x, y):
    return np.cos(x**2 + y**2)

x, y = np.meshgrid(x, y)
z = f(x, y)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

colorbar = fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)
colorbar.set_label('Function Value', fontsize=12)

ax.set_xlabel('X-axis', fontsize=12)
ax.set_ylabel('Y-axis', fontsize=12)
ax.set_zlabel('Z-axis', fontsize=12)
ax.set_title('$f(x, y) = \\cos(x^2 + y^2)$', fontsize=14)

plt.show()
```

7. Besh xil mahsulot uchun savdo ma'lumotlarini aks ettiruvchi vertikal chiziqli diagramma yarating: ['Mahsulot A', 'Mahsulot B', 'Mahsulot C', 'Mahsulot D', 'Mahsulot E']. Har bir mahsulot uchun sotish qiymatlari [200, 150, 250, 175, 225]. Diagrammani sarlavha, o'q yorliqlari va turli bar ranglari bilan moslashtiring.

```python
values = [200, 150, 250, 175, 225]
items = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
colors = ['red', 'blue', 'green', 'black', 'brown']

plt.bar(items, values, 0.5, color=[i for i in colors])
plt.title('Bar Chart')
plt.xlabel('Products')
plt.ylabel('Numbers')
plt.show()
```
