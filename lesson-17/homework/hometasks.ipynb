```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2 -4*x + 4

x = np.linspace(-10, 10, 10000)
y = f(x)

plt.plot(x, y, c='skyblue', label='$f(x)=x^2 -4x + 4$')
plt.title('Square Function')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.legend()
plt.show()
```

```python
def cosinus(n):
    return np.cos(n)

def sinus(n):
    return np.sin(n)

x = np.linspace(0, 2*np.pi, 100)
cos = cosinus(x)
sin = sinus(x)

plt.plot(x, cos, c='r', label='$f(x)=cos(x)$')
plt.plot(x, sin, c='y', label='$f(x)=sin(x)$')
plt.title('Bar Chart')
plt.xlabel('Cos(x)')
plt.ylabel('Sin(x)')
plt.legend()
plt.show()
```

```python
def f1(x):
    return x**3

def f2(x):
    return np.sin(x * np.pi)

def f3(x):
    return np.e ** x

def f4(x):
    return np.where(x >= 0, np.log10(x + 1), None)

n = np.linspace(-2, 2, 100)
x = np.arange(20)

plt.subplot(2, 2, 1)
plt.plot(x, f1(x), c='k', label='$f(x)=x^3$')
plt.title('Kubik funksiya')
plt.xlabel('X')
plt.ylabel('$f(x)=x^3$')
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(n, f2(n), c='r', label='$f(x) = sin(x)$')
plt.title('Sinus funksiya')
plt.xlabel('X')
plt.ylabel('$f(x) = sin(x)$')
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, f3(x), c='g', label='$f(x)=e^x$')
plt.title('Exponeneta funksiya')
plt.xlabel('X')
plt.ylabel('$f(x)=e^x$')
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, f4(x), c='b', label='$f(x)=log(x + 1$')
plt.title('Logarfimik funksiya')
plt.xlabel('X')
plt.ylabel('$f(x)=log(x + 1$')
plt.legend()

plt.show()
```

```python
x = np.random.normal(0, 10, 100)
y = np.random.normal(0, 10, 100)

plt.scatter(x, y, color='skyblue', marker='*')
plt.title('Scatter Graph')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
n = np.random.normal(0, 1, 1000)

plt.hist(n, bins=30, color='#243678', edgecolor='white', alpha=0.6)
plt.title('Histogramma')
plt.xlabel('Items')
plt.ylabel('Values')
plt.show()
```

```python
x = np.arange(-5, 5, 0.01)
y = np.arange(-5, 5, 0.01)

def f(x, y):
    return np.cos(x**2 + y**2)

x, y = np.meshgrid(x, y)
z = f(x, y)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
colorbar = fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)
ax.set_xlabel('X-axis', fontsize=12)
ax.set_ylabel('Y-axis', fontsize=12)
ax.set_zlabel('Z-axis', fontsize=12)
ax.set_title('$f(x, y)=cos(x^2 + y^2)$', fontsize=14)

plt.show()
```

```python
values = [200, 150, 250, 175, 225]
items = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
colors = ['red', 'blue', 'green', 'black', 'brown']

plt.bar(items, values, 0.5, color=[i for i in colors])
plt.title('Bar Chart')
plt.xlabel('Products')
plt.ylabel('Numbers')
plt.show()
```

```python
x = ['Category A', 'Category B', 'Category C']
y1 = np.array([20, 20, 30])
y2 = np.array([30, 35, 45])
y3 = np.array([14, 53, 31])
y4 = np.array([26, 45, 25])

plt.bar(x, y1, color='r', label='1900 years')
plt.bar(x, y2, bottom=y1, color='b', label='1950 years')
plt.bar(x, y3, bottom=y1+y2, color='g', label='2000 years')
plt.bar(x, y4, bottom=y1+y2+y3, color='k', label='2025 years')
plt.title('Stacked Bar Chart')
plt.xlabel('Category')
plt.ylabel('Periods')
plt.legend()
plt.show()
```
