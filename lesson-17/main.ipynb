### Bar Chart

```python
colors = ['red', 'blue', 'green', 'yellow']
numbers = [6, 13, 9, 5]

plt.bar(colors, numbers, 0.2, color=[i for i in colors], )
plt.title('Basic Bar Chart')
plt.xlabel('Colors')
plt.ylabel('Numbers')
plt.show()
```

```python
colors = ['red', 'blue', 'green', 'yellow']
numbers = [6, 13, 9, 5]
digit = [11, 8, 5, 17]
n = np.arange(len(colors))

plt.xticks(n, colors)
plt.bar(n-0.1, numbers, 0.2, color='b', label='Blue Color')
plt.bar(n+0.1, digit, 0.2, color='g', label='Green Color')
plt.title('Grouped Bar Chart')
plt.xlabel('Color')
plt.ylabel('Numbers')
plt.legend()
plt.show()
```

```python
colors = ['red', 'blue', 'green', 'yellow']
numbers = [6, 13, 9, 5]
digit = [11, 8, 5, 17]
n = np.arange(len(colors))

plt.xticks(n, colors)
plt.bar(colors, numbers, 0.2, color='b', label='Blue Color')
plt.bar(colors, digit, 0.2, bottom=numbers, color='g', label='Green Color')
plt.title('Grouped Bar Chart')
plt.xlabel('Color')
plt.ylabel('Numbers')
plt.legend()
plt.show()
```

### Histogramma

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randn(1000)

plt.hist(data, bins=20, color='skyblue', edgecolor='white')
plt.title('Basic Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
np.random.seed(42)
data1 = np.random.normal(0, 1, 1000)
data2 = np.random.normal(0, 1, 1000)

plt.hist(data1, bins=20, alpha=0.5, label='Dataset1', color='blue', edgecolor='black')
plt.hist(data2, bins=20, alpha=0.5, label='Dataset2', color='red', edgecolor='black')
plt.title('Histogram with Multiple Datasets')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

### 3D Images

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

ax.plot3D(x, y, z, c='g')
ax.set_title('3D line plot geeks for geeks')
plt.show()
```

$$ f(x) = 2x^2 - 3x + 1 $$

```python
x = np.linspace(-5, 6, 10000)

def f_x(x):
    return 2*x**2 - 3*x +1

y = f_x(x)

plt.plot(x, y, label='$f(x)=2x^2 - 3x + 1$', c='g')
plt.title('Parabola')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.legend()
plt.show()
```

```python
x = np.linspace(-5, 6, 10000)
plt.figure(figsize=(10, 3))

def f_x(x):
    return 3 * np.sin(3*x + np.pi)

y = f_x(x)

plt.plot(x, y, label='$f(x)=3x + sin(3x + pi)$', c='b')
plt.title('Parabola')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.legend()
plt.show()
```

```python
x = np.linspace(-5, 6, 10000)

def f_x(x):
    return np.where(x<0, x**2, np.sin(x))

y = f_x(x)

plt.plot(x, y, c='b')
plt.title('Function')
plt.xlabel('X')
plt.ylabel('Y', rotation=0)
plt.show()
```

```python
def f(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

x = np.linspace(-5, 6, 1000)
y = np.linspace(-5, 6, 1000)

x, y = np.meshgrid(x, y)
z = f(x, y)



fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(x, y, z, cmap='viridis')

```
